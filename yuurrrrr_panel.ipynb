```python
def __str__(self):
    newick = self._newick(self.root)
    newick.extend(';')
    return ''.join(newick)

def _newick(self, node, newick=[]):
    if len(node.children) > 0:
        newick.extend('(')
        for child in node.children:
            newick = self._newick(child, newick)
        if newick[-1] == ',':
            newick = newick[:-1]
        newick.extend(')')
    else:
        newick.extend(node.name)
        newick.extend(',')
    return newick
```

```python
import panel as pn
import pandas as pd
import hvplot.pandas

data = [
    "Number1 d:Bacteria,p:Actinobacteriota,c:Actinobacteria,o:Micrococcales,f:Micrococcaceae",
    "Number2 d:Bacteria,p:Proteobacteria,c:Alphaproteobacteria,o:Rhizobiales,f:Methyloligellaceae,g:uncultured",
    "Number5 d:Bacteria,p:Firmicutes,c:Bacilli,o:Bacillales,f:Bacillaceae,g:Bacillus"
]

parsed_data = []

for entry in data:
    parts = entry.split(' ')
    number = parts[0]
    taxonomy_info = {item.split(':')[0]: item.split(':')[1] for item in parts[1:]}
    parsed_data.append({"Number": number, "Taxonomy": taxonomy_info})


df = pd.DataFrame(parsed_data)

# Create a list of unique taxonomy levels
taxonomy_levels = set(item for sublist in df['Taxonomy'].apply(lambda x: x.keys()).tolist() for item in sublist)

# Create a dropdown widget for taxonomy levels
taxonomy_dropdown = pn.widgets.Select(options=list(taxonomy_levels), value='p', name='Taxonomy Level')

def update_bar_chart(taxonomy_level):
    counts = df['Taxonomy'].apply(lambda x: x.get(taxonomy_level, 'unknown')).value_counts()
    bar_chart = counts.hvplot.bar(title=f'Counts of {taxonomy_level.capitalize()}').opts(
        hooks=[lambda x: x.opts(tools=['tap'], active_tools=['tap'])]
    )
    return bar_chart


@pn.depends(taxonomy_level=taxonomy_dropdown.param.value)
def update_app(taxonomy_level):
    return update_bar_chart(taxonomy_level)

# Link the selections between the bar chart and the dropdown
linked_selections = pn.panel(df.hvplot.table(selected=[]))  # Create an empty table to hold the linked selections
pn.links.link_selections(update_bar_chart('p'), linked_selections).servable()
```

```python
taxonomy_levels = dict(zip(["d", "p", "c", "o", "f", "g", "s"], ["domain", "phylum", "class", "order", "family", "genus", "species"]))


class Node():
    def __init__(self, name, tax_level=None, count=None):
        self.name = name
        self.level = tax_level
        self.children = []
        self.parent = None
        self.zotus = []

    def add_child(self, child):
        self.children.append(child)
        child.parent = self

    def set_count(self, count):
        self.count = count

    def __repr__(self):
        first_zotu = "None" if len(self.zotus) == 0 else self.zotus[0]
        parent_name = "None" if self.parent is None else self.parent.name
        return f"\nNode name={self.name}; parent={parent_name}; children={len(self.children)}; zotus={len(self.zotus)}; {self.zotus}"

class Tree():
    def __init__(self):
        self.root = Node("root", "root")
        self.nodes = dict()
        self.nodes["root"] = self.root

    def add_lineages_file(self, file):
        with open(file) as lin_file:
            for line in lin_file:
                try:
                    self.add_lineage_str(line.strip())
                except Exception as ex:
                    print(ex, '\n', line)

    def add_lineage_str(self, lineage_str):
        #Zotu1  d:Bacteria,p:Actinobacteriota,c:Actinobacteria,o:Micrococcales,f:Micrococcaceae
        zotu, lineage = lineage_str.split("\t")
        lineage = lineage.split(",")
        parent = self.root
        for node_str in lineage:
            level, name = node_str.split(":")
            if name not in self.nodes:
                node = Node(name, taxonomy_levels[level])
                parent.add_child(node)
                node.parent = parent
                parent = node
                self.nodes[name] = node
            else:
                parent = self.nodes[name]
        #last node gets zotu annotated
        self.nodes[name].zotus.append(zotu)

    def get_nodes_as_string(self):
        return [n.__repr__().replace('\n','') for n in self.nodes.values()]

    # def __repr__(self):
    #     return f"Tree: {''.join([n.__repr__() for n in self.nodes.values()])}"

if __name__ == "__main__":
    tree = Tree()
    tree.add_lineages_file('/Users/cheyennebrouwer/Documents/23-24/Kwartaal_2/Data_Dashboards/Bioclear_earth/ASV_taxonomy.txt')

    print (tree.get_nodes_as_string())
```

```python

data = my_data_read

children = []
parents = []

for entry in data:
    parts = entry.split(';')
    if len(parts) >= 2:
        node_name = parts[0].split('=')[1].strip()
        parent = parts[1].split('=')[1].strip() if parts[1].split('=')[1].strip() != 'None' else ''
        children.append(node_name)
        parents.append(parent)

fig = go.Figure(go.Treemap(
    labels=children,
    parents=parents,
    root_color="lightgrey"
))

fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))
fig.show()
```
